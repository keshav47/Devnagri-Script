```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from tkinter import *
from keras.models import load_model
import numpy as np
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
from matplotlib.pyplot import imshow


class Root(Tk):
    def __init__(self):
        super(Root,self).__init__()
        self.title("DEVNAGRI COMPILER")
        self.minsize(640,400)
        
        self.labelFrameNew = ttk.LabelFrame(self,text='Train The Model')
        self.labelFrameNew.grid(column=0,row=1,padx=20,pady=20)
        
        
        self.labelFrame = ttk.LabelFrame(self,text='Input the Image to be Compiled')
        self.labelFrame.grid(column=0,row=4,padx=20,pady=20)
        
        self.button_1()
        
        self.button_3()
        
        self.button_4()
        
    
       
        
    def button_1(self):
        self.button_1 = ttk.Button(self.labelFrameNew,text='Test Set Accuracy',command= self.LoadDataSet)
        self.button_1.grid(column =1,row=2)
        
    def button_3(self):
        self.button_3 = ttk.Button(self.labelFrame,text='Browse A File',command= self.fileDialog)
        self.button_3.grid(column =1,row=5)
    
    def button_4(self):
        self.button_4 = ttk.Button(self.labelFrame,text='Process the image',command= self.devnagriModel)
        self.button_4.grid(column =1,row=7)
        
    
    
    
    def LoadDataSet(self):
        X_test=np.load("X_test.npy")
        Y_test=np.load("Y_test.npy")
        devnagri_model = load_model("devnagri_params.h5")
        preds = devnagri_model.evaluate(X_test, Y_test)
        accuracy=('the accuracy on the test set is =',preds[1]*100,'%')
        self.accuracyLabel= ttk.Label(self.labelFrameNew,text='')
        self.accuracyLabel.grid(column=1,row=3)
        self.accuracyLabel.configure(text=accuracy)
        
        
        
    def fileDialog(self):
        self.filename = filedialog.askopenfilename(initialdir='/',title="Select A file",filetype= (("jpeg","*.png"),("All Files","*.*")))
        self.label = ttk.Label(self.labelFrame,text="")
        self.label.grid(column=1,row=6)
        self.label.configure(text= self.filename)
    
    def devnagriModel(self):
        devnagri_model = load_model("devnagri_params.h5")
        img = image.load_img(self.filename, target_size=(32, 32))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        Prediction={0:'ka',1:'kha',2:'ga',3:'gha',4:'kna',5:'cha',6:'chha',7:'ja',8:'jha',9:'yna',10:'taamatar',11:'thaa',12:'daa',13:'dhaa',14:'adna',15:'tabala',16:'tha',17:'da',18:'dha',19:'na',20:'pa',21:'pha',22:'ba',23:'bha',24:'ma',25:'yaw',26:'ra',27:'la',28:'waw',29:'motosaw',30:'petchiryakha',31:'patalosaw',32:'ha',33:'chhya',34:'tra',35:'gya',36:'0',37:'1',38:'2',39:'3',40:'4',41:'5',42:'6',43:'7',44:'8',45:'9'}
        a=np.argmax(devnagri_model.predict(x))
        Result=('The Image loaded is = ',Prediction[a])
        self.resultLabel= ttk.Label(self.labelFrame,text='')
        self.resultLabel.grid(column=1,row=8)
        self.resultLabel.configure(text=Result)
    
    
    
    
    
if __name__ =='__main__':
    root = Root()
    root.mainloop()
    
    
        
        
        
```
